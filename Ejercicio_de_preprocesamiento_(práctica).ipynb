5. Realizar una codificación one-hot cualquier característica nominal

```python
# selector de categoricas
categoricas = make_column_selector(dtype_include= 'object')
```

```python
categoricas(X_train)
```

```python
# selector de numericas
numericas = make_column_selector(dtype_include= 'number')
```

```python
numericas(X_train)
```

```python
# sex, smoker, region
# podemos hacer 3 encoder individuales, uno para cada atributo
# o usar un transformador de columna

# creamos el subconjunto de las categoricas
X_train_categoricas = X_train[categoricas(X_train)]
X_test_categoricas = X_test[categoricas(X_test)]
```

```python
X_train_categoricas.head()
```

```python
X_test_categoricas.head()
```

```python
# instanciar one hot
onehot = OneHotEncoder(sparse = False, handle_unknown='ignore')
```

```python
# entrenar el oh
onehot.fit(X_train_categoricas)
```

```python
# tranformar los datos
X_train_encoded = onehot.transform(X_train_categoricas)
X_test_encoded = onehot.transform(X_test_categoricas)
```

```python
X_train_encoded
```

```python
# convertir en dataframe con una nueva columna por cada categoria de cada categorica de la codificacion
onehot_coulmns_name = onehot.get_feature_names_out(X_train_categoricas.columns)
```

```python
X_train_categoricas.columns
```

```python
onehot_coulmns_name
```

```python
X_train_encoder = pd.DataFrame(X_train_encoded, columns = onehot_coulmns_name)
```

```python
X_train_encoder.head()
```

```python
X_test_encoder = pd.DataFrame(X_test_encoded, columns = onehot_coulmns_name)
```
